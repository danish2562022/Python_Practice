
# What is Exploratory Data Analysis(EDA)? and why we do it?

EDA is an approach to get an insights about the data or we can say to find out the characteristics of data.


Now a days, many big organization want to make the strategic decisions based on the data they have and EDA plays a most important role to study about the DATA


## EDA of Iris Flower Dataset

```
# Importing libraries
import pandas as pd   # Pandas library is for manipulating or creating dataframe
import seaborn as sns  # Seaborn library is for plotting
import matplotlib.pyplot as plt # Matplotlib is for plotting
import numpy as np   # Numpy library perform mathematical operations on data
```

```
# Reading Data
iris=pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```
# printing first 5 rows of dataset
iris.head()
```

```
# Finding number of datapoints

print("Number of Datapoints: ",iris.shape[0])
```

```
# List of columns names

print("Columns present in Dataset: ", iris.columns)
```

### What is Scatter Plot?

In scatter plot each datapoint plotted as a point on X-Y axis and axis represents the value of features.


### 2-D Scatter Plot

```
sns.set_style("whitegrid")
sns.FacetGrid(iris,hue="species",size=6).map(plt.scatter,"sepal_length","sepal_width").add_legend()
plt.show()
```

### 3-D Scatter Plot

```
import plotly.express as px
df = px.data.iris()
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width',
              color='species')
fig.show()
```

### Pairplot - This is pairwise sactter plot

```
sns.pairplot(iris,hue="species",size=3)
plt.show()
```

From the pairplot, We can reach on following conclusion:

1. Setosa is linearly seperable(from a plane/line) from the other categories of flower.
2.petal_length of setosa flower is lesser than other two flowers and pdf of petal_length are less overlapped compare to other features



### As we saw, petal_length is most important features to differentiate setosa, virginica and versicolor,so we perform more analysis over petal_length features.

### Histogram, PDF, CDF

```
iris_setosa=iris.loc[iris["species"]=="setosa"]
iris_virginica=iris.loc[iris["species"]=="virginica"]
iris_versicolor=iris.loc[iris["species"]=="versicolor"]

plt.plot(iris_setosa['petal_length'],np.zeros_like(iris_setosa['petal_length']),'o')
plt.plot(iris_virginica['petal_length'],np.zeros_like(iris_virginica['petal_length']),'o')
plt.plot(iris_versicolor['petal_length'],np.zeros_like(iris_versicolor['petal_length']),'o')
```

```
plt.close()

sns.FacetGrid(iris,hue="species",height=7).map(sns.distplot,"petal_length").add_legend()
plt.show()
```

```
sns.FacetGrid(iris, hue="species", size=7) \
   .map(sns.distplot, "petal_width") \
   .add_legend();
plt.show();
```

```
sns.FacetGrid(iris, hue="species", size=7) \
   .map(sns.distplot, "sepal_length") \
   .add_legend();
plt.show();
```
